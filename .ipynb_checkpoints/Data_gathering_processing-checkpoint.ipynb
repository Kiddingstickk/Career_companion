```python
import pandas as pd
import random
from faker import Faker

fake = Faker()

# --------------------------
# Roles and skills
# --------------------------
role_skills = {
    "Java Developer": ["Java", "Spring", "Hibernate", "SQL", "REST APIs", "Git", "Maven"],
    "Testing": ["Selenium", "JUnit", "TestNG", "Manual Testing", "Bug Tracking", "Automation"],
    "DevOps Engineer": ["Docker", "Kubernetes", "Jenkins", "CI/CD", "Terraform", "AWS", "Linux"],
    "Python Developer": ["Python", "Django", "Flask", "Pandas", "NumPy", "REST APIs", "Git"],
    "Web Designing": ["HTML", "CSS", "JavaScript", "React", "Bootstrap", "UI/UX", "Photoshop"],
    "Hadoop": ["Hadoop", "Spark", "Hive", "Pig", "HDFS", "MapReduce", "Java"],
    "Blockchain": ["Solidity", "Ethereum", "Smart Contracts", "Truffle", "Web3.js", "Node.js"],
    "ETL Developer": ["Informatica", "Talend", "SQL", "Data Warehousing", "Python", "ETL Testing"],
    "Data Science": ["Python", "R", "Machine Learning", "Statistics", "Pandas", "NumPy", "Matplotlib", "SQL"],
    "Database": ["SQL", "MySQL", "Oracle", "PostgreSQL", "T-SQL", "Stored Procedures", "Backups"],
    "DotNet Developer": [".NET", "C#", "ASP.NET", "MVC", "SQL", "Entity Framework", "Visual Studio"],
    "Automation Testing": ["Selenium", "Appium", "TestNG", "JUnit", "CI/CD", "Python", "Java"],
    "Network Security Engineer": ["Firewalls", "IDS/IPS", "VPN", "Network Monitoring", "Penetration Testing", "Wireshark"],
    "SAP Developer": ["SAP ABAP", "SAP Fiori", "SAP HANA", "SAP Modules", "SQL", "Data Migration"]
}

# --------------------------
# Single resume generator
# --------------------------
def generate_synthetic_resume(role, role_skills, variation=True):
    name = fake.name()
    email = fake.email()
    location = f"{fake.city()}, {fake.state_abbr()}"
    skills = role_skills[role].copy()
    
    if variation:
        remove_count = random.randint(0, max(1, len(skills)//3))
        skills_to_remove = random.sample(skills, remove_count)
        for s in skills_to_remove:
            skills.remove(s)
        extra_skills_pool = [s for r, sks in role_skills.items() if r != role for s in sks]
        extra_skills = random.sample(extra_skills_pool, random.randint(0, 2))
        skills.extend(extra_skills)
    
    description = f"{name} is an experienced {role} with skills in {', '.join(skills)}. Adept at delivering quality work in fast-paced environments."
    
    experience_list = []
    num_jobs = random.randint(1, 3)
    for _ in range(num_jobs):
        job_title = role
        company = fake.company()
        start_year = random.randint(2010, 2020)
        end_year = start_year + random.randint(1, 5)
        responsibilities = f"Worked on projects involving {', '.join(random.sample(skills, min(len(skills), 3)))}."
        experience_list.append(f"{job_title} at {company} ({start_year}-{end_year}): {responsibilities}")
    
         # Education
        degree = random.choice(["B.Sc", "B.Tech", "M.Sc", "M.Tech", "MBA"])
        institution = f"{fake.company()} University"
        graduation_year = random.randint(2008, 2022)
        education = f"{degree} from {institution}, Class of {graduation_year}"

    
    num_projects = random.randint(1,2)
    projects = []
    for i in range(num_projects):
        techs = ", ".join(random.sample(skills, min(len(skills), 3)))
        projects.append(f"Project {i+1}: Developed using {techs}")
    
    return {
        "Name": name,
        "Email": email,
        "Location": location,
        "Role": role,
        "Skills": ", ".join(skills),
        "Summary": description,
        "Experience": " | ".join(experience_list),
        "Education": education,
        "Projects": " | ".join(projects)
    }

# --------------------------
# Generate dataset with real-world + synthetic 50-50
# --------------------------
def generate_combined_dataset(real_csv_path, synthetic_per_role=None):
    real_df = pd.read_csv(real_csv_path)
    num_real = len(real_df)
    
    # Calculate how many synthetic per role
    roles_count = len(role_skills)
    if synthetic_per_role is None:
        synthetic_per_role = (num_real // 2) // roles_count  # 50% synthetic
    
    synthetic_data = []
    for role in role_skills.keys():
        for _ in range(synthetic_per_role):
            synthetic_data.append(generate_synthetic_resume(role, role_skills))
    
    synthetic_df = pd.DataFrame(synthetic_data)
    
    # Combine and shuffle
    combined_df = pd.concat([real_df, synthetic_df], ignore_index=True)
    combined_df = combined_df.sample(frac=1).reset_index(drop=True)
    
    return combined_df

# --------------------------
# Example usage
# --------------------------
if __name__ == "__main__":
    synthetic_data = []
    synthetic_per_role = 150  # generate 20 resumes per role
    for role in role_skills.keys():
        for _ in range(synthetic_per_role):
            synthetic_data.append(generate_synthetic_resume(role, role_skills))
    
    combined_df = pd.DataFrame(synthetic_data)
    print(f"Total synthetic resumes generated: {len(combined_df)}")
    combined_df.to_csv("synthetic_resumes.csv", index=False)
```

```python
df_new = pd.concat([df[['Category' , 'Resume']], df2[['Category', 'Resume']]], ignore_index=True)
```

```python
df_new['Category'].value_counts()
```

```python
sampled_dfs = []

for cat in colms:
    df_cat = df_new[df_new['Category'] == cat].sample(n=174, random_state=42)
    sampled_dfs.append(df_cat)

df_sampled = pd.concat(sampled_dfs, ignore_index=True)

print(df_sampled['Category'].value_counts())
```

```python
import re

def clean_resume(text):
    text = text.encode('ascii', errors='ignore').decode('utf-8')
    
    text = re.sub(r'[^\w\s]', '', text)

    text = re.sub(r'\d+', '', text)

    text = re.sub(r'\s+', ' ', text)

    return text.strip()

df_sampled['Resume'] = df_sampled['Resume'].apply(clean_resume)
```

```python
df_sampled['Resume'] = df_sampled['Resume'].str.lower()
```
