```python
df2['difficulty'] = df2['difficulty'].map({'Easy' : 1 , 'Medium' : 2 , 'Hard' : 3})
df2['source_type'] = df2['source_type'].map({'Open-Ended' : 1 , 'Behavioral' : 2 , 'Situational' : 3 , 'STAR-Based' : 4})
df2['experience'] = df2['experience'].map({'fresher' : 1 , 'intern' : 2 , '1 year' : 3 , '2 years' : 4 , '3 years' : 5 , '4 years' : 6 , '5 years' : 7 , '6 years' : 8 , '7+ years' : 9 })
df2['role'] = df2['role'].map({'QA Analyst' : 1 , 'Product Manager' : 2 , 'HR Specialist' : 3 , 'UX Designer' : 4 , 'Software Engineer' : 5 , 'DevOps Engineer' : 6 , 'Data Scientist' : 7 , 'Marketing Associate' : 8})
df2['category'] = df2['category'].map({'Adaptability' : 1 , 'Culture Fit' : 2 , 'Motivation' : 3 , 'Work Style' : 4 , 'Team Collaboration' : 5 , 'Leadership' : 6 , 'Career Goals' : 7 , 'Conflict Resolution' : 8} )
```

```python
df2
```

```python
all_words = [word for row in df2['keywords'] for word in row]
word_to_id = {word: idx+1 for idx, word in enumerate(set(all_words))}
word_to_id
```

```python
df2['keywords'] = df2['keywords'].apply(lambda words: [word_to_id[word] for word in words])
```
